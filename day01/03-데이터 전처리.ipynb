## #07. 두 개의 이상의 데이터프렘임 병합하기 

### 1) 행 단위 병합

#### a) 샘플 DataFrame 만들기 

```python
df_top = DataFrame({'국어':[90,82], '수학':[81,76]}, index=['민철', '철수'])
df_top
```

```python
df_middle = DataFrame({'국어':[70,62], '영어':[77,68]}, index=['영민', '정수'])
df_middle
```

```python
df_bottom = DataFrame({'영어':[70,88], '과학':[81,76]}, index=['민철', '태영'])
df_bottom
```

#### b) A에 B를 이어붙이기

- 단순히 세로로 병합만 수행한다.
- 데이터프레임간에 컬럼이 다르더라도 존재하지 않는 컬럼은 빈값(결측치)로 채워 넣는다.
- 병합하는 데이터프레임간의 컬럼이 서로 다르다면 `sort=False`를 적용해야 경고가 발생하지 않는다.
    - `sort` 파라미터는 컬럼 이름을 정령하는 옵션

```python
df_a = df_top.append([df_middle, df_bottom], sort=False)
df_a
```

#### b) A, B, C 를 병합하기 

`append()`함수와 동일한 결과 
- 단순히 세로 방향으로 덧붙이는 개념이므로 인덱스가 중복되는 경우도 발생한다.
- sort 파라미터를 생략하거나 True로 지정한 경우 컬럼 순서가 이름순으로 정렬된다.
- (중요) 병합되는 데이터프레임간에 컬럼이 서로 다르다면 sort=False 파라미터 필수 지정.

```python
df_b = concat([df_top, df_middle, df_bottom], sort = False)
df_b
```

### 2) 열단위 병합

한 번에 두 개의 데이터프레임만 병합 가능함

#### a) 인덱스가 지정하지 않는 샘플 데이터 프레임 만들기

```python
df_left = DataFrame({'고개번호':[1001,1002,1003,1004],'이름':['철수','영희','민철','미영']})
df_left
```

```python
df_right = DataFrame({'고개번호':[1001,1002,1003,1005], '금액':[10000,20000,15000,5000]})
df_right
```

#### b) 두 데이터프레임의 공통 컬럼을 기준으로 병합하기 

양쪽 데이터프레임에 모두 존재하는 데이터만 보여준다.

> Database의 Inner Join과 같은 개념

```python
# 병합시 제외되는 항목들
# - df_left의 고객번호 1004번 데이터는 df_right의 고객번호 컬럼과 겹치지 않는다.
# - df_right의 고객번호 1005번 데이터는 df_left의 고객번호 컬럼과 겹치지 않는다.
merge(df_left, df_right)
```

#### c) 왼쪽 데이터프레임을 기준으로 오른쪽 프레임을 병합하기

`how=left` 파라미터는 왼쪽 데이터프레임의 모든 데이터를 보여준다.

> Database의 Left Outer Join과 같은 개념

```python
# - df_right의 고객번호 1005번 데이터는 df_left에 존재하지 않기 때문에 병합되지 않는다.
merge(df_left, df_right, how='left')
```

#### d) 오른쪽 데이터프레임을 기준으로 왼쪽 프레임을 병합하기

`how=right` 파라미터는 오른쪽 데이터프레임의 모든 데이터를 보여준다.

> Database의 Right Outer Join과 같은 개념

```python
# - df_left의 고객번호 1004번 데이터는 df_right에 존재하지 않기 때문에 병합되지 않는다.
merge(df_left, df_right, how='right')
```

#### e) 양쪽 데이터프레임을 모두 병합하기

`how=outer` 파라미터를 추가하면 공통 컬럼의 값이 한쪽에만 있어도 데이터를 보여준다.

> Database의 Full Outer Join과 같은 개념

```python
merge(df_left, df_right, how='outer')
```

#### g) 공통 컬럼이 두 개 이상 존재하는 경우의 병합

두 데이터프레임에서 이름이 같은 열은 모두 키가 된다.

```python
df_a = DataFrame({'고객명' :['민수','수영'],
                   '데이터':['20000','100000'],
                   '날짜': ['2018-01-01','2018-01-01']})
df_a
```

```python
df_b = DataFrame({'고객명':['민수','수영'], '데이터':['21세','20세']})
df_b
```

#### merge 처리하기 

`고객 컬럼`과 `데이터 컬럼`이 동시에 일치한느 데이터를 찾아 병합한다.

> 존재하지 않으므로 결과없음

```python
merge(df_a, df_b)
```

#### `on` 파라미터를 사용하여 기준 열 지정하기 

df_a의 `데이터`는 금액을 의미하지만 df_b의 `데이터`는 나이를 의미하므로 병합의 기준으로 사용되어서는 안된다.

```python
# 기준 열이 아니면서 이름이 같은 열에는  _x 또는 _y와 같은 접미사가 붙는다.
merge_tmp = merge(df_a, df_b, on=['고객명'])
merge_tmp
```

#### 컬럼이름을 수정하여 최종 결과 만들기

```python
merge_result = merge_tmp.rename(columns={'데이터_x':"금액", "데이터_y":"나이"})
merge_result
```

#### j) 왼쪽의 인덱스와 오른쪽의 컬럼을 기준으로 병합하기

```python
한국사 = DataFrame({'한국사':[87,91]}, index=['영희','철수'])
한국사
```

```python
세계사 = DataFrame({'세계사':[90,82], '이름':['영희','철수']})
세계사
```

##### 왼쪽에서는 인덱스 사용, 오른쪽에서는 `이름`열 사용

인덱스가 사라지고 모두 컬럼으로 지정된다.

```python
역사점수 = merge(한국사, 세계사, left_index=True, right_on=["이름"])
역사점수
```

#### `이름` 컬럼을 인덱스로 지정하고 삭제 하기 

> Pandas DataFrame에서는 인덱스가 그래프 표현에서의 x축을 담당하기 때문에 매우 중요한 역할을 한다고 볼수 있다.<br>
  그러므로 데이터프레임 생성시 인덱스가 부적절하면 적절한 전처리 과정을 거쳐 인덱스를 생성해야 한다.
  
현재 인덱스 추출  

```python
현재인덱스 = list(역사점수.index)
현재인덱스
```

새롭게 인덱스로 지정할 데이터 추출

```python
이름 = list(역사점수['이름'])
이름
```

`{before:after}` 형식의 딕셔너리로 구성

```python
indexDict = {}
for i, v in enumerate(이름):
    # `이름`의 i번째에 대응되는 항목을 `현재인덱스`에서 가져옴
    before = 현재인덱스[i];
    indexDict[before] = v
    
indexDict    
```

인덱스 변경

```python
역사점수.rename(index=indexDict, inplace=True)
역사점수
```

이름컬럼삭제

```python
역사점수.drop('이름', axis=1, inplace=True)
역사점수
```
